```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from sklearn.model_selection import train_test_split

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

from statsmodels.tsa.api import VAR, VARMAX
```

```python
df = pd.read_csv('alt_data/complete_final_data.csv')
```

```python
df['DATE'] = pd.to_datetime(df['DATE']).dt.to_period('M')
```

```python
df.set_index('DATE', inplace = True)
```

```python
df
```

```python
def interpret_dftest(dftest):
    dfoutput = pd.Series(dftest[0:3], index=['Test Statistic','p-value', 'Lag Used'])
    return dfoutput
```

```python
interpret_dftest(adfuller(df['avg_weeks_unemployed']))
```

```python
adfuller(df['avg_weeks_unemployed'])
```

```python
x
```

```python
target = ['target_mining_logging', 'target_construction', 'target_edu_health',
       'target_manufacturing', 'target_prof_business', 'target_gov',
       'target_leisure_hospitality']
```

```python
for column in df[target].columns:
    
    test = interpret_dftest(adfuller(df[column].dropna()))
    
    if test[1] > 0.01:
        
        test = interpret_dftest(adfuller(df['avg_weeks_unemployed'].diff(1).dropna()))
        df[f'{column}_first_diff']  = df[column].diff(1)
        
        if test[1] > 0.01:
            
            test = interpret_dftest(adfuller(df['avg_weeks_unemployed'].diff(1).diff(1).dropna()))
            df.drop(columns = [f'{column}_first_diff'], inplace = True)
            df[f'{column}_second_diff']  = df[column].diff(1).diff(1)
                       
```

```python
df.columns
```

```python
diff = ['target_mining_logging_second_diff',
       'target_construction_second_diff', 'target_edu_health_second_diff',
       'target_manufacturing_second_diff', 'target_prof_business_second_diff',
       'target_gov_second_diff', 'target_leisure_hospitality_second_diff']
```

```python
test_df = df.copy()
```

```python
test_df.dropna(inplace=True)
```

```python
test_df
```

```python
train, test = train_test_split(test_df,
                               test_size = 0.25,
                              shuffle=False)
```

```python
target = ['target_retail',
       'target_mining_logging', 'target_construction', 'target_edu_health',
       'target_manufacturing', 'target_prof_business', 'target_gov',
       'target_leisure_hospitality']
```

```python
model = VAR(train[diff],train.drop(columns = diff))
```

```python
model_lags= model.select_order(maxlags = 42)
```

```python
model_lags.summary()
```

```python
ts_model = model.fit(42,ic='aic')   
```

```python
ts_model.summary()
```

```python
ts_model.k_ar
```

```python
# Get the lag order
lag_order = ts_model.k_ar
print(lag_order)  #> 4
```

```python

# Input data for forecasting
forecast_input = test_df[diff].values[-lag_order:]
forecast_input
```

```python
fc = ts_model.forecast(forecast_input, lag_order, (test_df[-38:].drop(columns = diff)))
```

```python
df_forecast = pd.DataFrame(fc, index=df.index[-38:], columns=df[diff].columns + '_pred')
df_forecast
```

```python
def invert_transformation(data, pred):
    forecast = pred.copy()
    columns = data.columns
    for col in columns:
        forecast[str(col)+'_pred'] = data[col].iloc[-1] +   forecast[str(col) +'_pred'].cumsum()
    return forecast

output = invert_transformation(test[diff], df_forecast)
```

```python
new_df = pd.concat([test_df, output], ignore_index= True )
```
