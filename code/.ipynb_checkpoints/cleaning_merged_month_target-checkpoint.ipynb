# Data cleaning Pt. 3 - Merging monthly and target data together

```python
import numpy as np
import pandas as pd
```

```python
monthly = pd.read_csv('../data/monthly_data.csv')
targets = pd.read_csv('../data/all_targets_data.csv')
```

### Merging dataframes & setting datetime to `'DATE'` Column/setting at index

```python
monthly.merge(targets, how='left', left_index=True, right_index=True)
monthly['DATE'] = pd.to_datetime(monthly['DATE']).dt.to_period('M')
monthly.set_index('DATE', inplace=True)
```
